```python
class BankAccount:
  def __init__(self,owner,accno,balance):
    self.owner = owner
    self.accno = accno
    self.balance = balance

  def deposit(self,amount):
    self.balance = self.balance + amount
    print("Current Balance is",self.balance)

  def withdraw(self,amount):
    if amount <= self.balance:
      self.balance =self.balance - amount
      print("Current Balance is ",self.balance)
    else:
      print("Insufficient Balance")

cust1 =BankAccount("Aishu",23456,5000)
cust2 =BankAccount("Anu",24567,3000)

cust1.deposit(2000)
cust2.withdraw(7000)
cust1.withdraw(1000)


```

```python
region_name ="Sowthwest"
temperature_readings =[23.1,24.5,20.4,25.4,26.1] #in celsius
humidity_readings =[45.3,50.2,60.2,67.5,43.8] #in percentage

def whether_condition(humidity):
  if humidity > 60:
   return "Rainy"
  else :
    return "Sunny/Party Cloudy"

current_humidity = humidity_readings[-1]
condition = whether_condition(current_humidity)

print(condition)



```

```python
def calculate_average(data_list):
  if not data_list:
    return 0
  return sum(data_list)/len(data_list)

avg_temp = calculate_average(temperature_readings)
avg_humidity = calculate_average(humidity_readings)

print(f"Average temperature is {avg_temp}")
print(f"Average humidity is {avg_humidity}")
```
